```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

np.random.seed(42)
random.seed(42)

# Kathmandu zones
zones = [
    "Thamel", "Baneshwor", "Koteshwor", "Lalitpur",
    "Bhaktapur", "Balaju", "Kalanki", "Chabahil",
    "Maharajgunj", "Patan", "Boudha", "Kirtipur"
]

# Generate 6 months of hourly data
start = datetime(2024, 1, 1)
records = []

for day in range(180):  # 6 months
    current_date = start + timedelta(days=day)
    is_weekend = current_date.weekday() >= 5
    is_festival = random.random() < 0.05  # 5% chance

    for hour in range(24):
        for zone in zones:
            # Base demand by hour
            if 7 <= hour <= 9:      # morning rush
                base = random.randint(40, 80)
            elif 17 <= hour <= 20:  # evening rush
                base = random.randint(50, 90)
            elif 23 <= hour or hour <= 4:  # late night
                base = random.randint(5, 20)
            else:                   # normal hours
                base = random.randint(15, 45)

            # Modifiers
            if is_weekend: base = int(base * 1.3)
            if is_festival: base = int(base * 2.0)
            if zone == "Thamel": base = int(base * 1.5)

            # Weather effect
            weather = random.choice(
                ["Clear", "Clear", "Clear", "Rain", "Fog"])
            if weather == "Rain": base = int(base * 1.4)
            if weather == "Fog": base = int(base * 0.8)

            records.append({
                'date': current_date.strftime('%Y-%m-%d'),
                'hour': hour,
                'zone': zone,
                'demand': base,
                'weather': weather,
                'is_weekend': int(is_weekend),
                'is_festival': int(is_festival),
                'temperature': random.randint(10, 35),
                'day_of_week': current_date.weekday()
            })

df = pd.DataFrame(records)
df.to_csv(
    r'D:\projects\ride-demand-forecasting\kathmandu_rides.csv',
    index=False
)

print(f"✅ Generated {len(df):,} records!")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
print(df.head(10))
```
